# Read file

```python
import pandas as pd
df = pd.read_csv(r'data\supermarkets-commas.txt', index_col='ID')
df
```

# Geocoding

```python
from geopy.geocoders import ArcGIS
non = ArcGIS()
non.geocode("666 21st St, San Francisco, CA 94114")
```

## Set geocode address style

```python
df["Full Address"] = df["Address"] + ", " + df["City"] + ", " + df["State"]
df
```

```python
df["Location"] = df["Full Address"].apply(non.geocode)
df
```

```python
df["Latitude"] = df["Location"].apply(lambda loc: loc.latitude)
df["Longitude"] = df["Location"].apply(lambda loc: loc.longitude)
df
```
