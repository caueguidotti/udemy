# Read file - Indexing

```python
df_txt_header = pandas.read_csv(r'data/data.txt', header=None, names=['ID', 'Address', 'City', 'State', 'Country', 'Name', 'Employees'])
df_txt_header
```

```python
df_txt_header.columns = ['ID', 'Address', 'City', 'ZIP', 'Country', 'Name', 'Employees']
df_txt_header
```

```python
df_txt_header.set_index("ID", inplace=True, drop=False)
df_txt_header
```

```python
df_txt = df_txt_header
df_txt.loc[2:3, "City":"Country"]
```

```python
df_txt.loc[2, "Country"]
```

```python
df_txt.loc[:, "Country"].to_list()
```

```python
df_txt.iloc[0:2, [1, 3, 5]]
```

```python
df_txt.drop(1, axis=0)
```

```python
df_txt.drop(df_txt.iloc[:, [2, 4]].columns, axis=1)
```

```python
df_txt.drop(df_txt.loc[3:5, :].index, axis=0)
```

```python
df_txt['Continent'] = df_txt.shape[0]*['North America']
df_txt
```

```python
df_txt['FullAddress'] = df_txt['Address'] + ', ' + df_txt['City'] + ', ' + df_txt['Country'] + ', ' + df_txt['ZIP'] + ', ' + df_txt['Continent']
df_txt
```

```python
df_txt.drop(df_txt.iloc[:, [-2, -1]].columns, axis=1, inplace=True)
df_txt_t = df_txt.T
df_txt_t[7] = [7, 'My Address', 'My City', 'AB 12345', 'My Country', 'My Name', 0]
df_txt = df_txt_t.T
df_txt
```
